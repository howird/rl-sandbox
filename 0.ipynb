# Reinforcement Learning: Deep Q Learning
- Interplay between agent and env
- Agent observes state in env and acts upon that env
- Causes env to transition from one state to another
- Agent given reward based on that which tells the agent what is expected
- By finding corelation between states, actions, rewards, it learns to nav env
- env is what the agent reacts w/
- ex. robot that classifies faulty parts from conveyer belt based on their weight by separating them into 2 different bins, and a human checks them and gives the robot a score
    - state: position of the parts and the reward
    - not state: conveyer belt, bins, human
    - State/Env:
        - the classification (action) causes the state transition
        - State is just the reading of the weight sensor
        - Actions cause the new box to load -> new weight
        - The set of all possible states: state space = [min_weight, max_weight]
        - Human checker is irrelevant
    - Agent:
        - the software not the hardware
        - memory of the states, actions, rewards
        - decision making 
    - Actions:
        - discrete: good/bad/unkown
        - set of all actions -> action space
- Deep Q Learning is only for Discrete, for cts: Actor-Critic
- Is this just supervised learning?
    - in a limited info env

```python
import gym
import numpy as np
import matplotlib.pyplot as plt
```

```python
env = gym.make("FrozenLake-v1")
```

```python
n_games = 1000
win_pct = []
scores = []
```

```python
for i in range(n_games):
    done = False
    obs = env.reset()
    score = 0
    
    while not done:
        action = env.action_space.sample()
        obs, reward, done, trunc, info = env.step(action)
        score += reward
    scores.append(score)

    if i and i % 10 == 0:
        avg = np.mean(scores[:-10])
        win_pct.append(avg)
```

```python
plt.plot(win_pct)
```

```python
plt.plot(scores)
```
